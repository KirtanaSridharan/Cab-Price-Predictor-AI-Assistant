```python
import pandas as pd

from mapboxgl.utils import df_to_geojson
from mapboxgl.viz import CircleViz
```

```python
# Paste your Mapbox access token here
MAPBOX_ACCESS_TOKEN = "token"
```

```python
# Read the dataset
df = pd.read_csv("../dataset/test.csv") # Change to training data if needed

# Create the dataset for visualization
viz_data = df_to_geojson(df, lat='pickup_latitude', lon='pickup_longitude')

# Approx center of the map
nyc_center = (-74.0060, 40.7128)

# Visualize the pickup data points on a map using Mapbox
viz = CircleViz(viz_data, access_token=MAPBOX_ACCESS_TOKEN, center=nyc_center, zoom=10)
viz.show()
```
